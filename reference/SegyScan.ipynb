# SegyScan

``` python
SegyScan(fh, records, fs=None)
```

Representation of SEGY data grouped by shot.

## Parameters

| Name | Type | Description | Default |
|--------|---------------|----------------------------------------|----------|
| fh | FileHeader | File header shared by all scanned files. | *required* |
| records | list of ShotRecord | Collection of shot metadata describing trace segments. | *required* |

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [counts](#pysegy.SegyScan.counts) | Total number of traces for each shot. |
| [data](#pysegy.SegyScan.data) | Load data for all shots on first access. |
| [offsets](#pysegy.SegyScan.offsets) | First trace byte offset for every shot. |
| [paths](#pysegy.SegyScan.paths) | List of file paths corresponding to each shot. |
| [shots](#pysegy.SegyScan.shots) | Source coordinates for each shot including depth. |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [read_data](#pysegy.SegyScan.read_data) | Load all traces for a single shot. |
| [read_headers](#pysegy.SegyScan.read_headers) | Read only the headers for a single shot. |
| [summary](#pysegy.SegyScan.summary) | Header summaries for the `idx`-th shot. |

### read_data

``` python
read_data(idx, keys=None)
```

Load all traces for a single shot.

#### Parameters

| Name | Type | Description | Default |
|--------|--------------|----------------------------------------|-----------|
| idx | int | Index of the shot to read. | *required* |
| keys | Iterable\[str\] | Additional header fields to load with each trace. | `None` |

#### Returns

| Name | Type      | Description                                    |
|------|-----------|------------------------------------------------|
|      | SeisBlock | In-memory representation of the selected shot. |

### read_headers

``` python
read_headers(idx, keys=None)
```

Read only the headers for a single shot.

#### Parameters

| Name | Type | Description | Default |
|-------|--------------|-----------------------------------------|----------|
| idx | int | Shot index to read. | *required* |
| keys | Iterable\[str\] | Header fields to populate; by default all are read. | `None` |

#### Returns

| Name | Type                      | Description                            |
|------|---------------------------|----------------------------------------|
|      | list of BinaryTraceHeader | Parsed headers for the requested shot. |

### summary

``` python
summary(idx)
```

Header summaries for the `idx`-th shot.
