# ShotRecord

``` python
ShotRecord(
    path,
    coordinates,
    fileheader,
    rec_depth_key='GroupWaterDepth',
    segments=list(),
    summary=dict(),
    ns=0,
    dt=0,
    fs=None,
)
```

Information about a single shot location within a SEGY file.

## Attributes

| Name | Description |
|------------------------------------|------------------------------------|
| [dt](#pysegy.ShotRecord.dt) | int(\[x\]) -\> integer |
| [ns](#pysegy.ShotRecord.ns) | int(\[x\]) -\> integer |
| [rec_coordinates](#pysegy.ShotRecord.rec_coordinates) | Array of receiver coordinates for this shot. |
| [rec_depth_key](#pysegy.ShotRecord.rec_depth_key) | str(object=’’) -\> str |

## Methods

| Name | Description |
|------------------------------------|------------------------------------|
| [read_data](#pysegy.ShotRecord.read_data) | Load all traces for this shot. |
| [read_headers](#pysegy.ShotRecord.read_headers) | Read only the headers for this shot. |

### read_data

``` python
read_data(keys=None)
```
