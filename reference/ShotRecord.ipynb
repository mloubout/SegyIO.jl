# ShotRecord

``` python
ShotRecord(
    path,
    coordinates,
    fileheader,
    rec_depth_key='GroupWaterDepth',
    depth_key='SourceDepth',
    by_receiver=False,
    segments=list(),

    ns=0,
    dt=0,
    fs=None,
)
```
